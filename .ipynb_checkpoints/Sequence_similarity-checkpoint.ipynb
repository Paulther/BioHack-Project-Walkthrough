```python
import os
import subprocess

def run_mmseqs2_cluster(input_fasta, output_prefix, identity_threshold=0.9):
    """
    Run MMseqs2 to cluster sequences in a FASTA file based on sequence identity.

    Parameters:
        - input_fasta (str): Path to the input FASTA file.
        - output_prefix (str): Prefix for the output files generated by MMseqs2.
        - identity_threshold (float): Sequence identity threshold for clustering.
    """
    # Run MMseqs2 to cluster sequences
    mmseqs2_cmd = f"mmseqs easy-cluster {input_fasta} {output_prefix} {output_prefix}_tmp --min-seq-id {identity_threshold}"
    subprocess.run(mmseqs2_cmd, shell=True)

def main(input_fasta, output_prefix, identity_threshold=0.9):
    """Main function to cluster protein sequences using MMseqs2."""
    run_mmseqs2_cluster(input_fasta, output_prefix, identity_threshold)


    

identity_threshold = 0.3
input_fasta = CFG.home+"combined.fasta"
output_prefix = CFG.clusters+f'{str(100*identity_threshold)}p_clustering'

main(input_fasta, output_prefix, identity_threshold)
```

```python
def quick_split(prot_list, split_frac=0.8):
    random.seed(12345)
    '''
    Given a df of samples, randomly split indices between
    train and test at the desired fraction
    '''
    
    split = int(split_frac*len(prot_list))
    
    # shuffle indices
    idxs = list(range(len(prot_list)))
    random.shuffle(idxs)

    # split shuffled index list by split_frac
    train_idxs = idxs[:split]
    test_idxs = idxs[split:]
    
    # split dfs and return
    train_data = [prot_list[i] for i in train_idxs]
    test_data = [prot_list[i] for i in test_idxs]
        
    return train_data, test_data
```

```python
single = []
train_names = []
test_names = []
for i in group_size:
    result_df = df[df['Group'].str.contains(i, case=False, na=False)]
    names = []
    print(i)
    for idx, row in result_df.iterrows():
        names.append(row['Sequence'][0:4])
    if group_size[i] == 1:
        single.append(names[0])
    elif group_size[i] > 5:
        sampled = random.sample(names, 5)
        #train, test = quick_split(sampled)
        train, test = quick_split(names)
    else:
        train, test = quick_split(names)
    if group_size[i] != 1:
        train_names.append(train)
        test_names.append(test)
        
train, test = quick_split(single)
train_names.append(train)
test_names.append(test)

train_name = [item for sublist in train_names for item in sublist]
test_name = [item for sublist in test_names for item in sublist]
```
